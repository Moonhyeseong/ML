## K-Means Clustering

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 2  
x, y = make_blobs(random_state = 1)
mglearn.discrete_scatter(x[:,0], x[:,1], markers = 'o')
```

```python
#   .
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

```python
print(kmeans.labels_)
```

```python
print(kmeans.predict(x))
#print(kmeans.predict([[-10, -5]]))
```

```python
mglearn.discrete_scatter(x[:,0], x[:,1],kmeans.labels_, markers = 'o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],[0,1,2], markers = '^')
```

## K-Means Clustering 

```python
x_varied, y_varied = make_blobs(n_samples = 200 , cluster_std = [1.0, 2.5, 0.5], random_state = 170)
y_pred = KMeans(n_clusters = 3, random_state = 0).fit_predict(x_varied)

mglearn.discrete_scatter(x_varied[:,0], x_varied[:,1],y_pred)
plt.legend(["cluster 0" ,"cluster 1", "cluster 2"], loc = 'best')
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

```python
#    
X, y = make_blobs(random_state = 170, n_samples = 600)
rng = np.random.RandomState(74)

#   
transformation = rng.normal(size = (2,2))
X = np.dot(X , transformation)

#   KMeans 
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
y_pred = kmeans.predict(X)

#    
mglearn.discrete_scatter(X[:,0], X[:,1],kmeans.labels_, markers = 'o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],[0,1,2], markers = '^')
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

## Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
X, y = make_blobs(random_state = 1)
agg = AgglomerativeClustering(n_clusters = 3)
assigment = agg.fit_predict(X)

mglearn.discrete_scatter(X[:,0], X[:,1],assigment)
plt.legend(["cluster 0" ,"cluster 1", "cluster 2"], loc = 'best')
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
X, y = make_blobs(random_state = 0 , n_samples = 12)

dbscan = DBSCAN()
#dbscan = DBSCAN(min_samples = min_samples, eps = eps)
clusters = dbscan.fit_predict(X)
print(' :\n', clusters)

```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples = 200, noise = 0.05, random_state = 0)
plt.scatter(X[:, 0], X[:,1], edgecolors = 'black')
```

```python
#  k-means  
kmeans = KMeans(n_clusters = 2)
y_pred = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1],c = y_pred, edgecolor = 'k')
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

```python
# 0,  1    
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN()
clusters = dbscan.fit_predict(X_scaled)

#    .
plt.scatter(X_scaled[:,0], X_scaled[:,1],c = clusters, edgecolor = 'k')
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

## Dendrogram

```python
#Scipy  ward     
from scipy.cluster.hierarchy import dendrogram, ward

X, y = make_blobs(random_state = 0 , n_samples = 12)
#  X  ward  
#SciPy ward      
#    .
linkage_array = ward(X)

#     linkge_array    .
dendrogram(linkage_array)

#     
ax = plt.gca()
bounds = ax.get_xbound()
print(bounds)

ax.plot(bounds, [7.25, 7.25], '--', c='k')
ax.plot(bounds, [4, 4], '--', c='k')

ax.text(bounds[1], 7.25, 'two clusters', va = 'center', fontdict={'size': 15})
ax.text(bounds[1], 4, 'three clusters', va = 'center', fontdict={'size': 15})
plt.xlabel("Sample index")
plt.ylabel("cluster distance")
```
